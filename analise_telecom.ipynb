```python
# Importação das bibliotecas
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

tb = pd.read_csv("/content/drive/MyDrive/telecom_users.csv")

display(tb)
```

```python
# Limpando colunas que não serão Utilizadas

# axis -> 0 = linha; axis -> 1 = coluna
tb = tb.drop("Unnamed: 0", axis=1)
display(tb)
```

```python
# Informações da tabela

tb.info()
```

```python
# Tratando valores com type errado
tb["TotalGasto"] = pd.to_numeric(tb["TotalGasto"], errors="coerce")

# Tratando colunas que possuem todos os valores vazios
tb = tb.dropna(how="all", axis=1)
# Tratando linhas que possuem pelo menos 1 valor vazio
tb = tb.dropna(how="any", axis=0)
tb.info()
```

```python
# Verificando quantos clientes fizeram cancelamento
print(tb["Churn"].value_counts())
# verificação em porcentagem
print(tb["Churn"].value_counts(normalize=True).map("{:.1%}".format ))
```

```python
# Histograma para ver a relação com o churn
for coluna in tb.columns: 
  grafico = px.histogram(tb, x=coluna, color="Churn", text_auto=True)
  grafico.show()
```

```python
# Histograma para ver a relação com o churn usando o seaborn
for coluna in tb.columns:
    if coluna != "Churn":  # Ignorar a coluna "Churn"
        plt.figure()
        sns.histplot(data=tb, x=coluna, hue="Churn", multiple="stack", stat="count")
        plt.title(f"Histograma de {coluna} por Churn")
        plt.xlabel(coluna)
        plt.ylabel("Contagem")
        plt.show()
```
